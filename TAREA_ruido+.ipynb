```julia
using PyPlot
```

```julia
f1=1 #frec 1
f2=5 #frec 2
T=4 #tiempo de muestreo
sps=10 #muestras por segundo
t=collect(linspace(0,seg,seg*sps)) #vector de tiempo
y=cos.(2*π*t*f1) + cos.(2*π*t*f2);
```

```julia
using PyPlot
```

```julia
g=rand(length(y)); #Vector de "ruido"
y= cos.(2*π*t*f1) + sin.(2*π*t*f2) + g; #Función senoidal con ruido
```

```julia
using PyPlot
```

```julia
plot(t,y, "-g")
title("Dominio en el tiempo")
xlabel("Tiempo (s)")
ylabel("Amplitud")
```

```julia
F=abs.(fft(y,1)); # se emplea el algorito para la TDF llamado fft que se conoce como la transformada de Fourier rápida (fast fourirer transform) 
```
